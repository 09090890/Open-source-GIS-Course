### 介绍matplotlib绘图基础
介绍画板figure, 画布axes/subplot, 画布坐标系Axis，画布标题title等。


```python
import matplotlib.pyplot as plt
from osgeo import gdal
```

```python
path_img = 'data/Section_6/rsimg_chenggon_mndwi_thre025.tif'
```

```python
dset_mndwi = gdal.Open(path_img)
img_array = dset_mndwi.ReadAsArray()
print(img_array.shape)
```

```python
### 在画板上画图，默认画布数量为1
fig = plt.figure()
plt.imshow(img_array)
```

```python
### 自定义画布（方法1:add_axes()）
fig = plt.figure()
axe = fig.add_axes(rect=[0,0,0.7,0.7])
axe.imshow(img_array)
axe_inner = fig.add_axes(rect=[0.1,0.1,0.2,0.2])
axe_inner.imshow(img_array)
```

```python
### 自定义画布（方法2:add_subplot()）
fig = plt.figure(figsize=(10, 4))
axe_1 = fig.add_subplot(1,2,1)
axe_1.set_xlabel('column')
axe_1.set_ylabel('row')
axe_1.set_title('mndwi image of chenggong')
axe_1.imshow(img_array)
axe_2 = fig.add_subplot(1,2,2)
axe_2.imshow(img_array)
```
